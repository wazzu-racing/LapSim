```python
import pickle as pkl
from models import car_model
import numpy as np
from matplotlib import pyplot as plt

with open('nyoooooooom.pkl', 'rb') as f:
    vehicle = pkl.load(f)

radii = []
radii_amount = 20
for i in range(radii_amount):
    radii.append(4000/(-i+radii_amount))
car_angles = np.linspace(0, 10, 25)

AX_arrays = []
AY_arrays = []
torque_arrays = []

for r in radii:
    AX_row = []
    AY_row = []
    torque_row = []
    for angle in car_angles:
        data = vehicle.find_accurate_accel(r, angle * np.pi / 180)
        AX_row.append(data.AX)
        AY_row.append(data.AY)
        torque_row.append(data.torque)
    AX_arrays.append(AX_row)
    AY_arrays.append(AY_row)
    torque_arrays.append(torque_row)

legend = []
for i in AX_arrays:
    plt.plot(car_angles, i)
    legend.append(f'Radius: {radii[AX_arrays.index(i)]:.1f} in')
plt.title('Longitudinal Acceleration vs Car Angle')
plt.xlabel('Car Angle (degrees)')
plt.ylabel('Longitudinal Acceleration (in/s)')
plt.legend(legend)
plt.grid()
```
